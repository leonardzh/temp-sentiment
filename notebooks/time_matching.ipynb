```python
import pandas as pd

def process_csv(file_path):
    # Read the CSV file
    
    columns = ['id','tweet_created_at_int','lat','lon']
    df = pd.read_csv(file_path)[columns]
    # Assuming there's a Unix timestamp column named 'unix_timestamp'
    # Convert Unix timestamp to UTC datetime
    df['utc_datetime'] = pd.to_datetime(df['tweet_created_at_int'], unit='ms', utc=True)
    
    # Extract 'yyyymmdd' and 'hour'
    df['yyyymmdd'] = df['utc_datetime'].dt.strftime('%Y%m%d')
    df['hour'] = df['utc_datetime'].dt.hour
    
    # Select and rename columns as needed
    return df[columns+['yyyymmdd', 'hour']]  # Adjust 'your_columns_here' to match your data

def append_and_save(csv_files):
    appended_df = pd.DataFrame()
    
    # Process each file and append to the DataFrame
    for file in csv_files:
        processed_df = process_csv(file)
        print(file)
        appended_df = pd.concat([appended_df, processed_df], ignore_index=True)
    
    # Group by 'yyyymmdd' and save each group into a separate CSV
    for yyyymmdd, group in appended_df.groupby('yyyymmdd'):
        filename = f'../../../data/processed/pre_weather/id_utc_{yyyymmdd}.csv'
        group.to_csv(filename, index=False)
        print(f'Saved: {filename}')

# List of CSV file paths
csv_files = glob.glob("../../../data/processed/joined/*.csv")  # Update with your actual file paths

# Process and save
append_and_save(csv_files)
```
